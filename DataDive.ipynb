A quick look at the data shows that most columns have no missing data.  'Cabin', 'Embarked', and 'Age' are only  columns missing data.
```python
'''This function creates a new dataframe 
column is the column of interest that you want to group by
this works best with 'factor' data such as 'Sex' (male/female)
the df is the dataframe that is passed to the function
'''
def SurvivalRatesbyColumn(column,df):    
    Survived_df = pd.DataFrame(df.groupby(column)['Survived'].sum()).reset_index()
    Count_df = pd.DataFrame(df.groupby(column)['Survived'].count()).reset_index()
    Count_df['Total'] = Count_df['Survived']
    del Count_df['Survived']
    Survived_df = pd.merge(Survived_df,Count_df,on=column)
    Survived_df['Percent_Survived'] = Survived_df['Survived']/Survived_df['Total']
    return Survived_df
    
```

```python
#this function returns the factor variable 'Live' or 'Die'
#Makes it easier to see the results on a graph
def survival(judge):
    if judge == 0:
        decision = 'Decease'
    if judge == 1:
        decision = 'Live'
    return decision

titanic_df['LiveDecease'] = titanic_df['Survived'].apply(survival)
```

```python
#Returning the survival percentages by Sex
SurvivalbySex_df = SurvivalRatesbyColumn('Sex',titanic_df)
SurvivalbySex_df
```

```python
#Returning the survival percentages by Pclass
SurvivalbyPclass_df = SurvivalRatesbyColumn('Pclass',titanic_df)
SurvivalbyPclass_df
```

It would be interesting to categorize the Age groups into segments to compare the survival rates.
Here are some very loosely interpretted categories for age:
- Children < 15
- 15 <= YoungAdults < 25
- 25 <= Adults < 40
- 40 <= Elderly 

```python
#function to create factor variables categorical ages
def agefactor(age):
    if age < 15:
        aFactor = 'Children'
    if ((age >=15) & (age<25)):
        aFactor = 'YoungAdults'
    if ((age >=25) & (age<40)):
        aFactor = 'Adults'
    if (age >= 40):
        aFactor = 'Elderly'
    return aFactor

titanic_valid_Age['AgeFactor'] = titanic_valid_Age['Age'].apply(agefactor)
```
